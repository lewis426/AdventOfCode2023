```python
# opening the file in read mode 
my_file = open("input.txt", "r") 
  
# reading the file 
data = my_file.read() 
  
# split on ('\n')
data_list = data.split("\n") 
```

```python
card_dict = {}
for row in data_list:
    card_split = row.split(":")
    winning_cards = [cards for cards in card_split[1].split("|")][0]
    actual_cards = [cards for cards in card_split[1].split("|")][1]
    winning_nums = [num for num in winning_cards.split(" ") if num != ""]
    actual_nums = [num for num in actual_cards.split(" ") if num != ""]
    card_dict[card_split[0]] = [winning_nums, actual_nums]
```

```python
copies_dict = {int(card.split(" ")[-1]): 1 for card in card_dict.keys()}

for card, nums in card_dict.items():
    winning_count = 0
    card_num = int(card.split(" ")[-1])
    num_copies = copies_dict[card_num]
    if copies_dict[card_num] > 0:
        for num in nums[1]:
            if num in nums[0]:
                winning_count += 1

        print(f"card number is {card_num} and has {winning_count} winners with {num_copies} copy")

        copy_added = card_num + 1
        while winning_count > 0:
            next_copy = copy_added
            winning_count -= 1
            copies_dict[next_copy] += (1*num_copies)
            copy_added += 1
```
