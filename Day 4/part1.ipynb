```python
# opening the file in read mode 
my_file = open("input.txt", "r") 
  
# reading the file 
data = my_file.read() 
  
# split on ('\n')
data_list = data.split("\n") 
```

```python
card_dict = {}
for row in data_list:
    card_split = row.split(":")
    winning_cards = [cards for cards in card_split[1].split("|")][0]
    actual_cards = [cards for cards in card_split[1].split("|")][1]
    winning_nums = [num for num in winning_cards.split(" ") if num != ""]
    actual_nums = [num for num in actual_cards.split(" ") if num != ""]
    #print(winning_nums)
    #print(actual_nums)
    card_dict[card_split[0]] = [winning_nums, actual_nums]
```

```python
winning_dict = {}
for card, nums in card_dict.items():
    winning_count = 0
    for num in nums[1]:
        if num in nums[0]:
            if winning_count == 0:
                winning_count += 1
            else:
                winning_count *= 2
                #winning_count += 1
    
    winning_dict[card] = winning_count
    
```
