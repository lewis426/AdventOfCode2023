```python
def convert_to_ascii_sum(row):
    """
    Take an input text, and iterate over all chars by adding the ascii value and multiplying by 17.

        :params str row: Text pattern

        :returns int ascii_sum: Resultant sum of text pattern after applying rules
    """

    curr_val = 0
    for char in row:
        ascii_val = ord(char)
        # Increase curr_val
        curr_val += ascii_val
        curr_val *= 17
        curr_val = curr_val % 256

    return curr_val
```

```python
def split_pattern(row):
    """
    Take an input text, and split it by - or = to get a pattern and a lens length

        :params str row: Text pattern

        :returns str ascii_label_pattern: Resultant pattern ascii sum
        :returns int lens_len: Resultant lens length from pattern
        :returns str pattern_type: Determines whether - or = was used. each is dealt with differently
    """

    if "=" in row:
        pattern, lens_len = row.split("=")
        ascii_label_pattern = convert_to_ascii_sum(pattern)
        pattern_type = "="
        lens_len = int(lens_len)
    elif "-" in row:
        pattern, lens_len = row.split("-")
        ascii_label_pattern = convert_to_ascii_sum(pattern)
        pattern_type = "-"
    return pattern, ascii_label_pattern, lens_len, pattern_type
```

```python
# Build boxes with pattern and lens stores
pattern_dict = {}
for pattern in data:
    label, ascii_sum, lens_len, pattern_type = split_pattern(pattern)
    if pattern_type == "-":
        if ascii_sum in pattern_dict:
            if label in pattern_dict[ascii_sum][0]:
                idx = pattern_dict[ascii_sum][0].index(label)
                del pattern_dict[ascii_sum][1][idx]
                pattern_dict[ascii_sum][0].remove(label)
    elif pattern_type == "=":
        if ascii_sum in pattern_dict:
            if label in pattern_dict[ascii_sum][0]:
                idx = pattern_dict[ascii_sum][0].index(label)
                pattern_dict[ascii_sum][1][idx] = lens_len
            else:
                pattern_dict[ascii_sum][0].append(label)
                pattern_dict[ascii_sum][1].append(lens_len)
        else:
            pattern_dict[ascii_sum] = [[label],[lens_len]]
```

```python
pattern_dict
```

```python
lens_sum = []
box_lens_sum = []
for box_num, box_contents in pattern_dict.items():
    #print(box)#
    box_sum = 0
    slot = 0
    for label, lens_len in zip(box_contents[0],box_contents[1]):
        #print(label, lens_len, box_num+1, slot)
        slot += 1
        focusing_power = (box_num+1)*lens_len*slot
        box_sum += focusing_power
        print(f"{label}: box {box_num+1}, slot {slot}, focal length {lens_len}, focusing power of {focusing_power}")
        lens_sum.append(focusing_power)
    box_lens_sum.append(box_sum)
```
