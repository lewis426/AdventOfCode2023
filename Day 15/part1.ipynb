```python
import pandas as pd
import numpy as np
```

```python
# opening the file in read mode 
my_file = open("input.txt", "r") 
  
# reading the file 
data = my_file.read().split(',')
```

```python
data
```

```python
def convert_to_ascii_sum(row):
    """
    Take an input text, and iterate over all chars by adding the ascii value and multiplying by 17.

        :params str row: Text pattern

        :returns int ascii_sum: Resultant sum of text pattern after applying rules
    """

    curr_val = 0
    for char in row:
        ascii_val = ord(char)
        # Increase curr_val
        curr_val += ascii_val
        curr_val *= 17
        curr_val = curr_val % 256

    return curr_val
```

```python
pattern_list = []
for pattern in data:
    pattern_list.append(convert_to_ascii_sum(pattern))
```
