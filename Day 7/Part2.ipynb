```python
ranking_dict = {}
for card in cards:
    card_rank_val = []
    #print(card)
    card_rank_val.append(get_hand_type_rank(card))
    for char in card:
        card_rank_val.append(card_rank[char])
    ranking_dict[card] = card_rank_val
```

```python
def compare_decks(deck_ranks, other_deck_ranks):
    idx = 0
    decision = False
    while decision == False:
        if deck_ranks[idx] < other_deck_ranks[idx]:
            outcome = "win"
            decision = True
        elif deck_ranks[idx] > other_deck_ranks[idx]:
            outcome = "lose"
            decision = True
        else:
            idx += 1

    return outcome
```

```python
compare_decks([4,1,5,5,8,8], [4,1,5,6,7,8])
```

```python
sorted_ranking_dict = {k: v for k, v in sorted(ranking_dict.items(), key=lambda item: item[1], reverse=True)}
```

```python
final_ranking_order_dict = {}
rank = 1
for deck in sorted_ranking_dict.keys():
    final_ranking_order_dict[deck] = rank * bid_dict[deck]
    rank += 1
```
