```python
# opening the file in read mode 
my_file = open("input.txt", "r") 
  
# reading the file 
data = my_file.read() 
  
# split on ('\n')
data_list = data.split("\n") 
```

```python
bid_dict = {card: int(bid) for card, bid in [data.split(" ") for data in data_list]}
```

```python
cards = [data.split(" ")[0] for data in data_list]
```

```python
len(cards) == len(set(cards)) # shows all card combos are unique
```

```python
card_rank = {"A":1,"K":2,"Q":3, "J":4,"T":5,"9":6,"8":7,"7":8,"6":9,"5":10,"4":11,"3":12,"2":13}
```

```python
def get_hand_type_rank(cards):
    card_list = []
    match_dict = {}

    for card in cards:
        if card not in card_list:
            num_counts = cards.count(card)
            card_list.append(card_list)
            match_dict[card] = num_counts
    
    #print(match_dict)
    count_list =  list(match_dict.values())
    if 5 in count_list:
        rank = 1
    elif 4 in count_list:
        rank = 2
    elif (3 in count_list) and (2 in count_list):
        rank = 3
    elif 3 in count_list:
        rank = 4
    elif count_list.count(2) == 2:
        rank = 5
    elif 2 in count_list:
        rank = 6
    else:
        rank =7

    return rank
```

```python
get_hand_type_rank("AA345")
```

```python
ranking_dict = {}
for card in cards:
    card_rank_val = []
    card_rank_val.append(get_hand_type_rank(card))
    for char in card:
        card_rank_val.append(card_rank[char])
    ranking_dict[card] = card_rank_val
```

```python
def compare_decks(deck_ranks, other_deck_ranks):
    idx = 0
    decision = False
    while decision == False:
        if deck_ranks[idx] < other_deck_ranks[idx]:
            outcome = "win"
            decision = True
        elif deck_ranks[idx] > other_deck_ranks[idx]:
            outcome = "lose"
            decision = True
        else:
            idx += 1

    return outcome
```

```python
compare_decks([4,1,5,5,8,8], [4,1,5,6,7,8])
```

```python
sorted_ranking_dict = {k: v for k, v in sorted(ranking_dict.items(), key=lambda item: item[1], reverse=True)}
```

```python
final_ranking_order_dict = {}
rank = 1
for deck in sorted_ranking_dict.keys():
    final_ranking_order_dict[deck] = rank * bid_dict[deck]
    rank += 1
```
