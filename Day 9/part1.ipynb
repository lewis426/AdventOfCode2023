```python
# opening the file in read mode 
my_file = open("input.txt", "r") 
  
# reading the file 
data = my_file.read() 
  
# split on ('\n')
data_list = data.split("\n") 
```

```python
split_data = [data.split(" ") for data in data_list]
```

```python
def append_next_iteration(num_list):
    change_list = []
    num_list = [int(i) for i in num_list]

    for i in range(len(num_list)-1):
        change_list.append(int(num_list[i+1]) - int(num_list[i]))
    #print(change_list)
    if all(change == 0 for change in change_list):
        #print("use change list")
        stepped_list = change_list
    else:
        #print("use recursion")
        stepped_list = append_next_iteration(change_list)

    #print(change_list[-1])
    #print(stepped_list[-1])
    final_list = num_list + [num_list[-1] + stepped_list[-1]]
    #print(final_list)
    return final_list
```

```python
test_list = ["10",  "13",  "16",  "21",  "30", "45"]
```

```python
append_next_iteration(test_list)
```
