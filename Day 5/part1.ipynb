```python
map_dict = {}

for idx, map_list in enumerate(split_data):
    if idx == 0:
        seeds = [seed for seed in map_list[0].split(":")[1].split(" ") if seed != ""]
        map_dict["Seed"] = seeds
    else:
        for idx_2, item in enumerate(map_list):
            if idx_2 == 0:
                name = map_list[idx_2][:-1]
                map_dict[name] = []
            else:
                map_dict[name].append([x for x in item.split(" ") if x!= ''])
```

```python
map_dict
```

```python
def mapping_func(mapping_list):
    output_list = []
    for map in mapping_list:
        inp = int(map[1])
        outp = int(map[0])
        rng = int(map[2])
        change = outp - inp
        upper_bound = inp+rng-1

        output_list.append([inp, upper_bound, change])

    return output_list
```

```python
seed_soil_list = mapping_func(map_dict["seed-to-soil map"])
soil_fert_list = mapping_func(map_dict["soil-to-fertilizer map"])
fert_water_list = mapping_func(map_dict["fertilizer-to-water map"])
water_light_list = mapping_func(map_dict["water-to-light map"])
light_temp_list = mapping_func(map_dict["light-to-temperature map"])
temp_hum_list = mapping_func(map_dict["temperature-to-humidity map"])
hum_loc_list = mapping_func(map_dict["humidity-to-location map"])
```

```python
seed_soil_list
```

```python
def convert_map(input, map_lists):
    output = int(input)
    for map_list in map_lists:
        #print(input)
        #print(map_list)
        
        if (int(input) >= map_list[0]) & (int(input) <= map_list[1]):
            output = int(input) + map_list[2]
        
    return output
```

```python
final_soil = []
final_fert = []
final_water = []
final_light = []
final_temp = []
final_hum = []
final_loc = []

for seed in map_dict["Seed"]:
    soil = convert_map(seed, seed_soil_list)
    fert = convert_map(soil, soil_fert_list)
    water = convert_map(fert, fert_water_list)
    light = convert_map(water, water_light_list)
    temp = convert_map(light, light_temp_list)
    hum = convert_map(temp, temp_hum_list)
    loc = convert_map(hum, hum_loc_list)

    final_soil.append(soil)
    final_fert.append(fert)
    final_water.append(water)
    final_light.append(light)
    final_temp.append(temp)
    final_hum.append(hum)
    final_loc.append(loc)
```
