```python
map_dict = {}

for idx, map_list in enumerate(split_data):
    if idx == 0:
        seeds = [seed for seed in map_list[0].split(":")[1].split(" ") if seed != ""]
        seed_w_range = [":".join(seeds[i:i+2]) for i in range(0, len(seeds), 2)]
        map_dict["Seed"] = [x.split(":") for x in seed_w_range]
    else:
        for idx_2, item in enumerate(map_list):
            if idx_2 == 0:
                name = map_list[idx_2][:-1]
                map_dict[name] = []
            else:
                map_dict[name].append([x for x in item.split(" ") if x!= ''])
```

```python
def mapping_func(mapping_list):
    output_list = []
    for map in mapping_list:
        inp = int(map[1])
        outp = int(map[0])
        rng = int(map[2])
        change = outp - inp
        upper_bound = inp+rng-1

        output_list.append([inp, upper_bound, change])

    return output_list
```

```python
seed_soil_list = mapping_func(map_dict["seed-to-soil map"])
soil_fert_list = mapping_func(map_dict["soil-to-fertilizer map"])
fert_water_list = mapping_func(map_dict["fertilizer-to-water map"])
water_light_list = mapping_func(map_dict["water-to-light map"])
light_temp_list = mapping_func(map_dict["light-to-temperature map"])
temp_hum_list = mapping_func(map_dict["temperature-to-humidity map"])
hum_loc_list = mapping_func(map_dict["humidity-to-location map"])
```

```python
def convert_map(input, map_lists):
    output = int(input)
    for map_list in map_lists:
        #print(input)
        #print(map_list)
        
        if (int(input) >= map_list[0]) & (int(input) <= map_list[1]):
            output = int(input) + map_list[2]
        
    return output
```

```python
final_soil = []
final_fert = []
final_water = []
final_light = []
final_temp = []
final_hum = []
final_loc = []


min_loc = False
for idx, seed_rngs in enumerate(map_dict["Seed"]):
    print(f'Seed range {idx+1} of {len(map_dict["Seed"])}')
    for seed in range(int(seed_rngs[0]),int(seed_rngs[0])+int(seed_rngs[1])):
        #print(seed)
        soil = convert_map(seed, seed_soil_list)
        fert = convert_map(soil, soil_fert_list)
        water = convert_map(fert, fert_water_list)
        light = convert_map(water, water_light_list)
        temp = convert_map(light, light_temp_list)
        hum = convert_map(temp, temp_hum_list)
        loc = convert_map(hum, hum_loc_list)

        if min_loc == False:
            min_loc = loc
        else:
            min_loc = min(min_loc, loc)
```

```python
# code by: https://github.com/morgoth1145/advent-of-code/blob/5311ed667714398cf02d9b2bc2a4e88f53f7b0dc/2023/05/solution.py

seed_ranges = map_dict["Seed"]

for g in data_list[1:]:
    print(g+"--")
    step_mapping = [tuple(map(int, l.split()))
                        for l in g.splitlines()[1:]]
    new_ranges = []
    for start, rng in seed_ranges:
        start = int(start)
        rng = int(rng)
        while rng != 0:
            found_match = False
            best_dist = rng

            for dst, src, length in step_mapping:
                #print(dst,src,length)
                # if in range
                if src <= start < src+length:
                    # Found a match
                    off = start - src # min seed num - min map num = how much over min map num is it
                    rem_length = min(length - off, rng) # map rng - amount over map num OR seed rng = remaining possible values, either until end of map rng or end of seed rng
                    new_ranges.append((dst+off, rem_length)) # add map dst + offset to get num in range incl req offset, then include length of possible other vals
                    start += rem_length # update seed start to what is left until end of range to try again
                    rng -= rem_length # update seed range so its reduced to what left in map rng
                    found_match = True
                    break
                else:
                    # elif seed below map lower
                    if start < src:
                        best_dist = min(src - start, best_dist) # min value is curr best dist or diff between map lower and seed lower

            if not found_match:
                handling_len = min(best_dist, rng) # min of curr best dist and seed rng
                new_ranges.append((start, handling_len)) # new range of seed lower and min rng
                start += handling_len # update new start to be higher to get to best dist
                rng -= handling_len # update new rng to be lower to get best new rng

    seed_ranges = new_ranges
```
