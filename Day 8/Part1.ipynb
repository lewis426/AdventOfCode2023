```python
import pandas as pd
import numpy as np
from ast import literal_eval
```

```python
# opening the file in read mode 
my_file = open("input.txt", "r") 
  
# reading the file 
data = my_file.read() 
  
# split on ('\n')
data_list = data.split("\n\n") 
```

```python
instructions = data_list[0]
data_split = data_list[1].split('\n')
```

```python
node_dict = {}
for data in data_split:
    start_node = data.split("=")[0].strip()
    node_instr = (data.split("=")[1].strip()[1:4],data.split("=")[1].strip()[6:9])
    node_dict[start_node] = node_instr
```

```python
node_dict['NQT']
```

```python
current_node = "AAA"
num_steps = 0
complete = False

while complete == False:
    for step in instructions:
        if step == "R":
            idx = 1
        elif step == "L":
            idx = 0
        num_steps += 1
        print(f"Step {num_steps}: {current_node} to {node_dict[current_node][idx]}")
        current_node = node_dict[current_node][idx]
        
        if current_node == "ZZZ":
            complete = True
            break
```
