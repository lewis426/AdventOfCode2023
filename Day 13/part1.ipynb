```python
all_vertical_matches = []
all_horizontal_matches = []

for data_idx, data in enumerate(data_list):
    data_dict = {}
    for row_idx, row in enumerate([data for data in data.split("\n") if data != ""]):
        data_dict[row_idx] = []
        for col in row:
            data_dict[row_idx].append(col)
    
    # Compare rows
    match = False
    stop = False
    row_match_idx = []
    idx = 0
    while  (stop == False):
        if idx == len(data_dict)-1:
            stop = True
        elif data_dict[idx+1] == data_dict[idx]:
            match = True
            row_match_idx.append(idx)
            idx += 1
        else:
            idx+=1        
    
    if row_match_idx:
        stop = False
    else:
        stop = True

    confirmed_row_matches = []
    for idx in row_match_idx:
        down_idx = idx+1
        up_idx = idx
        horizontal_counter = 0
        horizontal_match = True
        stop = False
        while stop == False:
            #print(down_idx, up_idx)
            #print(len(data_dict))
            if (up_idx>=0) & (down_idx <= len(data_dict)-1):
                if data_dict[down_idx] != data_dict[up_idx]:
                    horizontal_match = False
                    stop = True
                else:
                    horizontal_counter += 1
                    up_idx -= 1
                    down_idx += 1
            else:
                stop = True
        #print(up_idx)
        if up_idx < 0:
            addition = 0
        else:
            addition = up_idx + 1
        if horizontal_match == True:
            confirmed_row_matches.append(horizontal_counter+addition)

    if confirmed_row_matches:
        horizontal_matches = max(confirmed_row_matches)#*100
        print(f"Found horizontal match for chart {data_idx+1} with counter of {max(confirmed_row_matches)}")
        all_horizontal_matches.append(horizontal_matches)
    else:
        horizontal_matches = 0

    temp_dict = {}
    
    #print(data_dict)
    for col_idx in range(len(data_dict[0])):
        #print(f"COL: {col_idx}")
        temp_list = []
        next_list = []
        for row_idx in range(len(data_dict)):
            #print(f"ROW: {row_idx}")
            temp_list.append(data_dict[row_idx][col_idx])
        temp_dict[col_idx] = temp_list

    #print(temp_dict)
    
    match = False
    stop = False
    col_idx = 0
    #print(temp_dict)
    col_match_idx = []
    while (stop == False):
        if col_idx == len(temp_dict)-1:
            stop = True
        elif temp_dict[col_idx+1] == temp_dict[col_idx]:
            match = True
            col_match_idx.append(col_idx)
            col_idx += 1
            #print("Found Match")
        else:
            col_idx += 1
        #print(temp_dict[col_idx+1])
        #print(temp_dict[col_idx])
    if col_match_idx:
        stop = False
        #print(col_match_idx)
    else:
        stop = True
        
    confirmed_col_matches = []
    for idx in col_match_idx:
        #print(idx)
        down_idx = idx+1
        up_idx = idx
        vertical_counter = 0
        vertical_match = True
        stop = False
        while stop == False:
            if (up_idx>=0) & (down_idx <= len(temp_dict)-1):
                #print(up_idx, down_idx)
                #print(temp_dict[down_idx])
                #print(temp_dict[up_idx])
                if temp_dict[down_idx] != temp_dict[up_idx]:
                    vertical_match = False
                    stop = True
                else:
                    vertical_counter += 1
                    up_idx -= 1
                    down_idx += 1
            else:
                stop= True
        #print(up_idx)
        if up_idx < 0:
            addition = 0
        else:
            addition = up_idx + 1
        if vertical_match == True:
            confirmed_col_matches.append(vertical_counter+addition)

    if confirmed_col_matches:
        vertical_matches = max(confirmed_col_matches)
        print(f"Found vertical match for chart {data_idx+1} with counter of {max(confirmed_col_matches)}")
        all_vertical_matches.append(vertical_matches)
    else:
        vertical_matches = 0


print(f"Total Counter is: {sum(all_vertical_matches)+sum(all_horizontal_matches)*100}")
```
