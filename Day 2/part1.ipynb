```python
import pandas as pd
```

```python
# opening the file in read mode 
my_file = open("input.txt", "r") 
  
# reading the file 
data = my_file.read() 
  
# split on ('\n')
data_list = data.split("\n") 
```

```python
data_list[:5]
```

```python
max_red = 12
max_green = 13
max_blue = 14
check_dict = {"blue":max_blue, "red":max_red, "green":max_green}
```

```python
test_list = ["Game 1: 3 blue, 4 red; 1 red, 2 green, 6 blue; 2 green"]
```

```python
game_dict = {}
for game in data_list:
    game_split = game.split(":")
    game_key = game_split[0]
    round_split = [round_split.split(",") for round_split in game_split[1].split(";")]
    round_list = []
    for round in round_split:
        for item in round:
            item = item.strip()
            colour_split = item.split(" ")
            #print(colour_split)
            colour_key = colour_split[1]
            colour_num = colour_split[0]
            temp_dict = {colour_key:colour_num}
            round_list.append(temp_dict)
    game_dict[game_key] = round_list
```

```python
game_dict
```

```python
result_dict = {}
for game, checks in game_dict.items():
    game_id = game.strip().split(" ")[1]
    valid = True
    for pick in checks:
        # Check each pick and compare wih max num cubes for each colour
        for colour, num in pick.items():
            max_cubes = check_dict[colour]
            # override to say selection is invalid if pick is larger than num cubes available for any
            if int(num) > max_cubes:
                valid = False
    # Update results with validity after all iterations for each game checked
    result_dict[game_id] = valid
```

```python
# Num valid games
sum(1 for v in result_dict.values() if v == True)
```

```python
# Num invalid games
sum(1 for v in result_dict.values() if v == False)
```

```python
def filter_valid(pair):
    key, value = pair
    if value == True:
        return True  # keep pair in the filtered dictionary
    else:
        return False  # filter pair out of the dictionary
```

```python
filtered_results = dict(filter(filter_valid, result_dict.items()))
```

```python
# sum() performs summation
sum([int(key) for key in filtered_results.keys()])
```
